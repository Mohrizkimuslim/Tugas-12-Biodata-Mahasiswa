```python
import tkinter as tk
from tkinter import ttk

def tampilkan_data():
    nim = entry_nim.get()
    nama = entry_nama.get()
    prodi = entry_prodi.get()
    
    text_output = (
        "========== BIODATA MAHASISWA ==========\n\n"
        f"NIM          : {nim}\n"
        f"Nama         : {nama}\n"
        f"Program Studi: {prodi}"
    )
    
    output_area.config(state='normal')
    output_area.delete('1.0', tk.END)
    output_area.insert(tk.END, text_output)
    output_area.config(state='disabled')

def reset_data():
    entry_nim.delete(0, tk.END)
    entry_nama.delete(0, tk.END)
    entry_prodi.delete(0, tk.END)
    
    output_area.config(state='normal')
    output_area.delete('1.0', tk.END)
    output_area.config(state='disabled')

root = tk.Tk()
root.title("Aplikasi Biodata Mahasiswa")
root.geometry("450x380")
root.resizable(False, False)

frame_input = tk.LabelFrame(root, text="Input Data", padx=10, pady=10)
frame_input.pack(fill="x", padx=15, pady=10)

lbl_nim = tk.Label(frame_input, text="NIM", width=12, anchor="w")
lbl_nim.grid(row=0, column=0, pady=5, sticky="w")
entry_nim = tk.Entry(frame_input, width=40)
entry_nim.grid(row=0, column=1, pady=5)

lbl_nama = tk.Label(frame_input, text="Nama", width=12, anchor="w")
lbl_nama.grid(row=1, column=0, pady=5, sticky="w")
entry_nama = tk.Entry(frame_input, width=40)
entry_nama.grid(row=1, column=1, pady=5)

lbl_prodi = tk.Label(frame_input, text="Program Studi", width=12, anchor="w")
lbl_prodi.grid(row=2, column=0, pady=5, sticky="w")
entry_prodi = tk.Entry(frame_input, width=40)
entry_prodi.grid(row=2, column=1, pady=5)

frame_buttons = tk.Frame(root)
frame_buttons.pack(pady=5)

btn_tampilkan = tk.Button(frame_buttons, text="Tampilkan", width=12, command=tampilkan_data)
btn_tampilkan.grid(row=0, column=0, padx=5)

btn_reset = tk.Button(frame_buttons, text="Reset", width=12, command=reset_data)
btn_reset.grid(row=0, column=1, padx=5)

lbl_output_title = tk.Label(root, text="Output", anchor="w")
lbl_output_title.pack(fill="x", padx=15, pady=(10, 2))

output_area = tk.Text(root, height=7, bg="white", font=("Courier", 10))
output_area.pack(fill="both", padx=15, pady=5, expand=True)
output_area.config(state='disabled')

root.mainloop()
```
